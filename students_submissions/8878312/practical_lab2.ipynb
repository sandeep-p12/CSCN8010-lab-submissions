# Data Visualization Notebook

## In this notebook, we will explore various data visualizations using Matplotlib and Seaborn. We will create four different types of plots to illustrate different aspects of data.

### 1. Scatter Plot
A scatter plot is used to visualize the relationship between two continuous variables. In this example, we will plot the relationship between the "X" and "Y" variables.

```python
import matplotlib.pyplot as plt
import numpy as np

# Random data
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', label='Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot Example')
plt.legend()
plt.show()
```

### 2. Histogram
A histogram is used to visualize the distribution of a single variable. In this example, we will create a histogram to show the distribution of exam scores.

```python
import matplotlib.pyplot as plt
import numpy as np

# Random exam scores
np.random.seed(0)
scores = np.random.normal(70, 10, 100)

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(scores, bins=20, color='green', alpha=0.7)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')
plt.show()
```

### 3. Line Plot
A line plot is used to visualize the trends and patterns in time-series data. In this example, we will plot the monthly sales data over a year.

```python
import matplotlib.pyplot as plt
import numpy as np

# Monthly sales data
months = np.arange(1, 13)
sales = np.array([100, 120, 130, 110, 150, 140, 160, 180, 200, 190, 210, 220])

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(months, sales, marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.grid(True)
plt.show()
```

### 4. Box Plot
A box plot is used to visualize the distribution and spread of a dataset. In this example, we will create a box plot to compare the salaries of employees in different departments.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=data, palette="Set3")
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount ($)')
plt.title('Box Plot of Total Bill Amount by Day')
plt.show()
```
